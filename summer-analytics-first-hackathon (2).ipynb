```python
train_data = pd.read_csv('/kaggle/input/summer-analytics-mid-hackathon/hacktrain.csv')
test_data = pd.read_csv('/kaggle/input/summer-analytics-mid-hackathon/hacktest.csv')
train_data
test_data   # test data doesnt' contain the answers 
```

```python
train_data.drop(columns = ['ID'],inplace = True )
train_data
```

```python
si = SimpleImputer()
train_data.iloc[:,2:] = si.fit_transform(train_data.iloc[:,2:])
train_data
```

```python
le = LabelEncoder()
train_data['class'] = le.fit_transform(train_data['class'])
train_data
```

```python
x = train_data.drop(columns = ['class'])
y = train_data['class']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 42 , test_size = 0.2 , stratify = y)
x_train.isnull().sum()
```

```python
lr = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=10000
)
lr.fit(x_train,y_train)
y_pred =lr.predict(x_test)
```

```python
print(classification_report(
    y_test,
    y_pred,
    labels=list(range(len(le.classes_))),
    target_names=le.classes_
))
```

```python
ID = test_data['ID']
test_data.drop(columns = ['ID'],inplace  = True)

Y_predicted = lr.predict(test_data)
Y_predicted
```
